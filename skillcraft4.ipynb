```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.pyplot import figure
%matplotlib inline
```

```python
accidents_df = pd.read_csv('/content/accident.csv')
```

```python
print(accidents_df.head())
```

```python
print(accidents_df.isnull().sum())
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(x='Road_Type', data=accidents_df, palette='viridis')
plt.title('Distribution of Accidents by Road Condition')
plt.xlabel('Road Condition')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(x='Weather_Conditions', data=accidents_df, palette='viridis')
plt.title('Distribution of Accidents by Weather')
plt.xlabel('Weather')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(x='Time', data=accidents_df, palette='viridis')
plt.title('Distribution of Accidents by Time of Day')
plt.xlabel('Time of Day')
plt.ylabel('Count')
plt.show()

```

```python
plt.figure(figsize=(10, 5))
sns.countplot(x='Road_Type', hue='Weather_Conditions', data=accidents_df, palette='viridis')
plt.title('Accidents by Road Condition and Weather')
plt.xlabel('Road Condition')
plt.ylabel('Count')
plt.legend(title='Weather')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.scatterplot(x='Driver_Fatigue', y='Alcohol_Involved', hue='Road_Type', style='Weather_Conditions', s=100, palette='viridis', data=accidents_df)
plt.title('Accident Hotspots')
plt.xlabel('Driver_Fatigue')
plt.ylabel('Alcohol_Involved')
plt.legend(title='Road Condition & Weather')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(x='Number_of_Injuries', data=accidents_df, palette='viridis')
plt.title('Distribution of Accidents by Number of Injuries')
plt.xlabel('Number of Injuries')


```
